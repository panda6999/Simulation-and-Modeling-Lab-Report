```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# set random data from normal distributions
np.random.seed(42)

#sample 1
mean1 = 18
std_dev1 = 2
sample1 = np.random.normal(mean1, std_dev1, 50)

#sample 2
mean2 = 15
std_dev2 = 3
sample2 = np.random.normal(mean2, std_dev2, 50)
```

```python
#perform the two-sample t-test
t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)

print(f"t-statistic: {t_statistic:.2f}")
print(f"p-value: {p_value:.6f}")
```

```python
#conclusion
if p_value < 0.05:
    print("The means of the two samples are statistically different.")
else:
    print("There is no statistically significant difference between the means of the two samples.")

#visualization
plt.figure(figsize=(6, 4))
plt.bar(['sample 1', 'sample2'], [np.mean(sample1), np.mean(sample2)], color=['purple', 'black'], alpha=0.7)
plt.errorbar(['sample 1', 'sample2'], [np.mean(sample1), np.mean(sample2)], yerr=[np.std(sample1), np.std(sample2)], fmt='none', color='black', capsize=5)
plt.title ('Comparison of Sample Means')
plt.ylabel('Mean Vlaues')
plt.grid(True)
plt.show()
```
