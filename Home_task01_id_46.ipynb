```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vec1 = np.random.rand(15)
vec2 = np.random.rand(15)

print(f"vec1:\n{vec1}\nvec2:\n{vec2}")

vec1_flat = vec1.flatten()
vec2_flat = vec2.flatten()

x = np.arange(len(vec1_flat))
print(x)

#plot the data
plt.plot(x, vec1_flat, marker='o', linestyle='-', label = "VECTOR 1")
plt.plot(x, vec2_flat, marker='s', linestyle='--', label = "VECTOR 2")

#labels and title
plt.xlabel("index of vector")
plt.ylabel("Vlues of vectoe")
plt.title("line graph of flattened vector 1 and 2")
plt.legend()
plt.show()
```

```python
matrix = np.random.rand(5, 5)
print(f"Matrix:\n{matrix}")

sns.heatmap(matrix, annot=True, cmap="Blues", linewidths=.5)
plt.title("Heatmap of 5x5 Matrix")
plt.xlabel("Colums")
plt.ylabel("Rows")
plt.show()
```

```python
A = np.random.randint(0, 10, size=(4, 4))
B = np.random.randint(0, 15, size=(4, 4))

print(f"Matrix A:\n{A}")
print(f"Matrix B:\n{B}")

matrix_sum = A + B
matrix_diff = A - B
matrix_prod =np.dot(A, B)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# summation
print('The summation of them is:\n', matrix_sum)
axes[0].bar(range(1, 17), matrix_sum.flatten(), color='green')
axes[0].set_title('Summation')
axes[0].set_xlabel('index')
axes[0].set_ylabel('values')

# Difference
print('The difference of them is:\n', matrix_diff)
axes[1].bar(range(1, 17), matrix_diff.flatten(), color='purple')
axes[1].set_title('Difference')
axes[1].set_xlabel('index')
axes[1].set_ylabel('values')

# Product
print('The product of them is:\n', matrix_prod)
axes[2].bar(range(1, 17), matrix_prod.flatten(), color='blue')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('index')
axes[2].set_ylabel('Value')

plt.tight_layout()
plt.show()
```
